```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the newly uploaded dataset to inspect its structure and identify relevant columns
file_path_new = '/content/worldometer_coronavirus_summary_data.csv'

# Load the dataset
coronavirus_data = pd.read_csv(file_path_new)

# Display the first few rows to inspect the dataset and identify columns for correlation analysis
coronavirus_data.head()
```

```python
# Selecting relevant columns for correlation analysis
correlation_data = coronavirus_data[['total_confirmed', 'total_deaths', 'total_recovered',
                                     'active_cases', 'total_tests']]

# Calculating the correlation matrix
correlation_matrix = correlation_data.corr()

# Creating a heatmap to visualize the correlation between variables
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu', linewidths=0.5)
plt.title('Heatmap of COVID-19 Variables Correlation Across Countries', fontsize=16)
plt.show()
```
